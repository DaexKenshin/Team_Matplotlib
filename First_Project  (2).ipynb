```python

# List of unique countries
countries = ['Nepal', 'Mali', 'United States', 'United Kingdom', 'Canada', 'France']

# Group the DataFrame by 'Country' and count the occurrences of each social media platform
grouped = df.groupby('Country')['SocialMedia'].value_counts().unstack(fill_value=0)

# Iterate through the countries in the specified order and print the desired output
for country in countries:
    if country in grouped.index:
        social_media_counts = grouped.loc[country]
        sorted_counts = social_media_counts.sort_values(ascending=False)
        top_10_platforms = sorted_counts.head(10)
        print(f"{country}'s top 10 'SocialMedia")
        for platform, count in top_10_platforms.items():
            print(f"{platform} {count}")
        print("**************************************************")
```

```python
# Assuming you have a DataFrame named df with 'Country' and 'Income' columns

# Create a dictionary to store the statistics for each country
country_stats = {}

# Iterate through the list of countries
countries = ['Nepal', 'Mali', 'United States', 'United Kingdom', 'Canada', 'France']
for country in countries:
    # Filter the data for the current country
    country_data = df[df['Country'] == country]
    
    # Calculate the mean, max, and median
    mean = country_data['Salary_in_USD'].mean()
    max_value = country_data['Salary_in_USD'].max()
    median = country_data['Salary_in_USD'].median()
    
    # Store the statistics in the dictionary
    country_stats[country] = {'mean': mean, 'max': max_value, 'median': median}

# Print the statistics for each country
for country, stats in country_stats.items():
    print(f"{country}")
    print(stats['mean'])
    print(stats['max'])
    print(stats['median'])
    print("_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ ")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of unique countries
countries = ['Nepal', 'Mali', 'United States', 'United Kingdom', 'Canada', 'France']

# Group the DataFrame by 'Country' and count the occurrences of each social media platform
grouped = df.groupby('Country')['SocialMedia'].value_counts().unstack(fill_value=0)

# Iterate through the countries in the specified order and plot histograms
for country in countries:
    if country in grouped.index:
        social_media_counts = grouped.loc[country]
        sorted_counts = social_media_counts.sort_values(ascending=False)
        top_10_platforms = sorted_counts.head(10)

        # Create a bar chart (histogram) for the top 10 social media platforms
        plt.figure(figsize=(10, 6))  # Set the figure size
        plt.bar(top_10_platforms.index, top_10_platforms.values)
        plt.title(f"{country}'s Top 10 Social Media Platforms")
        plt.xlabel("Social Media Platform")
        plt.ylabel("Count")
        plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
        plt.tight_layout()

        plt.show()

        print("**************************************************")
```
